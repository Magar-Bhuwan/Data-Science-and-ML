# Machine Learning

Machine learning is a subset of artificial intelligence where algorithms identify patterns in data to make predictions or decisions without being explicitly programmed. It drives modern tech like recommendation systems, autonomous vehicles, and generative AI by training models on large datasets.

### Core Categories
Machine learning is primarily divided into four approaches, each suited for different types of problems:

- Supervised Learning: The model is trained on labeled data (inputs paired with known outputs). It learns the relationship to predict outcomes on new, unseen data. Common uses: Predicting house prices, image classification, and spam detection.

- Unsupervised Learning: The algorithm analyzes unlabeled data to find hidden structures, clusters, or patterns without human guidance. Common uses: Customer segmentation, anomaly detection, and data compression.

- Reinforcement Learning: The model learns through trial and error by receiving rewards or penalties for its actions, aiming to maximize a specific goal. Common uses: Robotics, game-playing AI, and self-driving cars.

- Semi-Supervised Learning: A hybrid approach utilizing a small amount of labeled data and a large amount of unlabeled data during training

### Common Algorithms
Depending on the task, developers use different statistical and mathematical models:

- Linear & Logistic Regression: Used for predicting continuous numerical values (e.g., forecasting sales) or determining probabilities (e.g., classification).

- Decision Trees & Random Forests: Tree-like structures used to classify data or predict values based on a series of feature-based rules.

- Neural Networks: Computing systems inspired by the human brain, forming the foundation of deep learning. They are highly effective for complex tasks like natural language processing and computer vision.

1. Supervised Learning

Supervised learning algorithms are generally categorized into two main types: 

- Classification: where the goal is to predict discrete labels or categories 
- Regression: where the aim is to predict continuous numerical values.

There are many algorithms used in supervised learning each suited to different types of problems. 

## Linear Regression

For simple linear regression (with one independent variable), the best-fit line is represented by the equation

y = mx + b

Where:

- y is the predicted value (dependent variable)
- x is the input (independent variable)
- m is the slope of the line (how much y changes when x changes)
- b is the intercept (the value of y when x = 0)

The best-fit line will be the one that optimizes the values of m (slope) and b (intercept) so that the predicted y values are as close as possible to the actual data points.

### Assumptions
- Linearity: The relationship between features and target is linear.
- Independence: Observations are independent of each other.
- Homoscedasticity: The residuals (errors) have constant variance
