# NumPy Operations

```python
import numpy as np
```

```python
# a = np.random.randint(1,6, (2,3))
# b = np.random.randint(6,10, (3,3))
# c = np.random.randint(1,9, (3,3))
a = np.array([[2, 3, 2],
            [5, 3, 1]])

b = np.array([[7, 9, 8],
                 [7, 6, 6],
                 [6, 6, 7]]),

c = np.array([[7, 2, 3],
              [4, 1, 5],
              [2, 4, 6]])

a, b, c
```

```python
np.round(np.sqrt(c), 2)
```

```python
np.pow(b, c)
```

```python
# .exp() function is used to calculate the exponential of all elements in the input array.

np.round(np.exp(a), 2)
```

```python
# .log10() function is used to calculate the logarithm of all elements in the input array to the base 10.
np.round(np.log10(c), 2)
```

```python
np.sum(a)
```

```python
np.sum(a, axis=0) # column wise sum
```

```python
# Returns the average of the array elements.
np.mean(a, axis=0)
```

```python
# Returns the standard deviation, a measure of the spread of a distribution, of the array elements.
np.std(a, axis=0)
```

```python
# Return the minimum of an array or minimum along an axis.
np.min(a, axis=0)
```

```python
np.max(a, axis=0)
```

```python
# condition 1: a == 1
# condition 2: a < 5
# Results: True if both conditions satisfied
np.logical_and(a == 1, a < 5)
```

```python
# True value where values are not 3 :  a != 3
# True only where a == 3 (reverse it) : logical_not
np.logical_not(a != 3)
```

```python
# condition 1: a > 1
# condition 2: a < 5
# Results: True if any one condition satisfied
np.logical_or(a > 5, a < 5)
```

```python
# reshape is use to re-size the array size (rows and columns)
a.reshape(2, 3)
```

```python
# flatten function makes 1d array list
a.flatten()
```

```python
# Transpose reverse the rows into columns and vice-versa
Transpose = a.T
Transpose
```

```python
# Accessing the element using the indexing. [row, column]
a
```

```python
a[-1,-1]
```

```python
# Slicing is use to get multiple elements (sub array) by (Start:End:Steps)
a[0:1]
```

```python
# Return the index (position) of maximum values (not the value itself)
a.argmax()
```

```python
# Return the index (position) of minimum values (not the value itself)
a.argmin()
```

```python
# Adds elements step-by-step
a.cumsum()
```

```python
# Multiplies elements step-by-step
a.cumprod()
```

```python
# Sorts array elements in ascending order
np.sort(a)
```

```python
# Finds indices where a condition is true
np.where(a > 0)
```

```python
# Returns unique (non-repeated) values
np.unique(a)
```
