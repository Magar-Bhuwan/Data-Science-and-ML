🔹 1. Array Creation Methods

Used to create NumPy arrays:

 - np.array() → create array from list

 - np.zeros() → array filled with 0

 - np.ones() → array filled with 1

 - np.arange() → range of values

 - np.linspace() → evenly spaced values

 - np.random.rand() → random numbers

🔹 2. Shape & Reshaping Methods

Control array structure:

 - arr.shape → get dimensions

 - arr.reshape() → change shape

 - arr.flatten() → convert to 1D

 - arr.T → transpose

🔹 3. Mathematical Operations

Perform calculations:

np.add(), np.subtract(), np.multiply(), np.divide()
np.sqrt(), np.power()
np.sum(), np.mean(), np.std()

🔹 4. Indexing & Slicing

Access elements:

 - arr[i] → single element

 - arr[start:end] → slicing

 - arr[:, 1] → column selection

🔹 5. Aggregation Methods

Summarize data:

 - arr.sum() → total

 - arr.min(), arr.max()

 - arr.argmax(), arr.argmin()

🔹 6. Joining & Splitting

Combine or break arrays:

 - np.concatenate()

 - np.vstack(), np.hstack()

 - np.split()

🔹 7. Sorting & Searching

 - np.sort() → sort array

 - np.where() → find condition

 - np.unique() → unique values

🔹 8. Copy & View

 - arr.copy() → new independent array

 - arr.view() → reference to original


```python
import numpy as np
```

```python
# Create a 3x4 array of random numbers
arr = np.random.rand(3, 4)
arr
```

```python
# .shape - returns a tuple of the dimensions of the array
arr.shape
```

```python
# .ndim - returns the number of dimensions of the array
arr.ndim
```

```python
# .dtype - returns the data type of the array
arr.dtype
```

```python
# .size - returns the total number of elements in the array
arr.size
```

```python
# Transpose of an array
arr_T = arr.T
arr_T.shape
```

```python
# Create a 3x4 array of random integers between 2 and 5
arr = np.random.randint(2,5, (3,4))
arr
```

```python
# Reshape the array to 2x3x2
arr.reshape(2, 3, -1)
```

```python
# .flatten - returns a 1D array of the elements in the array.
flat_arry = arr.flatten()
flat_arry
```

```python
# arr.sum(axis=0) # column wise sum
# arr.sum(axis=1) # row wise sum
# total sum of all elements in the array
arr.sum()
```

```python
arr.sum(axis=0)
```

```python
arr.sum(axis=1) 
```

```python
# row wise product of all elements in the array
arr.prod(axis = 1) 
```

```python
# cumulative sum along the column
arr.cumsum(axis =0) 
```

```python
# .max - returns the maximum value in the array
arr.max()
```

```python
# column wise minimum
arr.min(axis=0) 
```

```python
# column wise index of maximum value
arr.argmax(axis=0) 
```
