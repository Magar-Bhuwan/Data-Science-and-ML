# 1. Element Opeations

# Arithmetic Operations

```python
# b + c
np.add(b,c) # element wise addition of two arrays
```

```python
# b - c
np.subtract(b,c) # element wise subtraction of two arrays
```

```python
# b * c
np.multiply(b,c) # element wise multiplication of two arrays
```

```python
# b / c
np.round(np.divide(b,c), 2) # element wise division of two arrays
```

```python
# modulo
np.mod(b,c) # element wise modulo of two arrays
# b % c
```

```python
# a ** 2 # element wise exponentiation of an array with a scalar value
np.pow(a, 2) # element wise exponentiation of an array with a scalar value
# np.pow(b,c) # element wise exponentiation of two arrays
```
