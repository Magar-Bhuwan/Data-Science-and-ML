```python
import pandas as pd

data = pd.read_csv("DNA_Dataset_Normalized.csv")

data.head()
```

```python
data.Class.value_counts()
```

```python
X = data.drop("Class", axis=1).to_numpy()
Y = data["Class"].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, Y, test_size = 0.2, stratify=Y, random_state=999
)

log_res = LogisticRegression(max_iter=1000)
log_res.fit(X_train, y_train)

y_pred = log_res.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=25)

X_reduced = pca.fit_transform(X)
```

```python
X_reduced.shape
```

```python
sum(pca.explained_variance_ratio_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, Y, test_size = 0.2, stratify=Y, random_state=999
)

log_res = LogisticRegression(max_iter=500)
log_res.fit(X_train, y_train)

y_pred = log_res.predict(X_test)

accuracy_score(y_test, y_pred)
```
