## Creating NumPy Array

NumPy Array is also know as ndarray.
It is a powerful N-Dimensional gride of values used for numerical and scientific computing in python.
It has characteristics: Homogeneity, Fixed Size, Performance, etc.
NumPy Array Attributes:
 - ndim --> The number of dimensions in rank.
 - shape --> A tuple representing the size of each dimension (Rows & Columns)
 - size --> The total number of elements in the Array.
 - dtype --> The data types of elements (int64, float32)

```python
list1 = [1, 2, 3, 4, 5]
print(list1)
```

```python
# we can convert a list to a numpy array using the np.array() function. 
# This allows us to perform various operations on the array efficiently.

array1 = np.array(list1)
print(type(array1))
print(len(array1))
print(array1)
```

```python
# shape of the array use to total number of rows and columns in the array
array1.shape
```

```python
# ndim is use to find the number of dimensions in the array
array1.ndim
```

```python
# we can also create a 2D array (matrix) using a list of lists. Each inner list represents a row in the matrix.
list2 = [
    [1, 2, 3, 4],
    [5, 6, 7, 8]

]
print(list2)
```

```python
# we can convert the list of lists to a 2D numpy array using the np.array() function. 
# This will create a 2D array (matrix) where each inner list becomes a row in the matrix.
array2 = np.array(list2)
print(array2)
```

```python
# shape of the 2D array will give us the number of rows and columns in the matrix. 
# In this case, it will return (2, 4) which means there are 2 rows and 4 columns in the array.
array2.shape
```

```python
# ndim is used to find the number of dimensions in the array
array2.ndim
```

## Accessing an Array
Indexing and Slicing NumPy Arrays
NumPy arrays can be indexed and sliced in several ways depending on the number of dimensions.

Indexing:

Indexing follows a simple rule that we may have encountered before. Let's give it a try.
- 1D Array: array_name[idx] (Similar to Python list)
- 2D Array: array_name[row_idx, col_idx] (Similar to Matrix)
- ND Array: array_name[dim1_idx, dim2_idx, ..., dimN_idx]

OR

- 2D Array: array_name[row_idx][col_idx]
- ND Array: array_name[dim1_idx][dim2_idx] ... [dimN_idx]

Slicing:
Slicing follows the same rule as indexing. The only difference is that for each dimension we provide it with a range of indeces and sometimes (optional) step size (similar to arange() function discussed above).

- 1D Array: array_name[idx1:idx2:step]
- 2D Array: array_name[row_idx1:row_idx2:step1, col_idx1:col_idx2:step2]
- ND Array: array_name[dim1_idx1:dim1_idx2:step1, ..., dimN_idx1:dimN_idx2:stepN]

```python
# 1D array
a = np.array([1,2,3,4,5])
```

```python
# we can access elements in the array using indexing.
a[-3]
```

```python
# we can also access a range of elements in the array using slicing.
# slicing syntax is [start:stop:step]. If we omit start, it defaults to the beginning of the array. If we omit stop, it defaults to the end of the array. If we omit step, it defaults to 1.
a[:3:]
```

```python
# 2D array
b = np.random.randint(0, 20, size=(3, 4))
b
```

```python
# we can access elements in a 2D array using row and column indices. 
# The syntax is array[row_index, column_index].
b[1,0]
```

```python
# we can also access the last element in the array using negative indexing.
b[-1, -1]
```

```python
# we can also access elements in the array using negative indexing.
b[-3]
```

```python
# Slicing 2D Array
# Syntax: array[row_start:row_stop:row_step, column_start:column_stop:column_step]
b[0::2, 2::]
```

```python
# Syntax: array[row_start:row_stop:row_step, column_start:column_stop:column_step]
b[:2, 1]
```

```python
# 3D array (tensor)
# syntax: np.random.randint(low, high, size=(block of array, rows, columns))
c = np.random.randint(0, 20, size=(3, 3, 4))
c
```

```python
# we can access elements in a 3D array using three indices: block index, row index, and column index.
c[0,1,2]
```

```python
# 20 colorful images
# Syntax: np.random.randint(low, high, size=(number of images, color channels, height, width))
d = np.random.randint(0, 20, size=(20, 3, 10, 10))
d
```

```python
# d[-1] will give us the last image in the array, 
# which is a 3D array of shape (3, 10, 10) representing the color channels and dimensions of the image.
d[-1]
```
