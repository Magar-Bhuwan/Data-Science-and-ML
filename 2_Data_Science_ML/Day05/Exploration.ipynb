# Correlation and Covariance

In Python, covariance measures the direction of the linear relationship between two variables, while correlation measures both the direction and the strength in a standardized scale from -1 to 1

```python
titanic_df[["Age", "Fare", "SibSp"]].corr()
```

```python
titanic_df[["Age", "Fare"]].cov()
```
