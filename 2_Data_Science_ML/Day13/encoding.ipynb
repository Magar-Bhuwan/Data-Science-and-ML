```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

titanic_df = pd.read_csv('update_titanic_data.csv')
```

```python
titanic_df.head()
```

```python
titanic_df.drop(
['Name', 'initials'], axis = 1, inplace = True)
titanic_df.head()
```

# Label Encoding

### Nominal Data
Data Without Natural Order/ Rank

```python
titanic_df.Gender
```

#### One Hot Encoding

```python
# Pandas for post processing
pd.get_dummies(
    titanic_df, columns=['Gender'], dtype=int
)
```

# Another Method using Scikit_learn for preprocessing

```python
onehot_encoder = OneHotEncoder(sparse_output=False)
```

```python
encode_gender = onehot_encoder.fit_transform(titanic_df[['Gender']])
encode_gender
```

```python
onehot_encoder.categories_
```

```python
onehot_encoder.inverse_transform(
    [[1,0],
     [0,1]]
)
```

```python
df = pd.DataFrame(
    encode_gender, columns=onehot_encoder.categories_[0], dtype=int
)
titanic_df = pd.concat([titanic_df,df], axis=1)
titanic_df.head()
```

# Lable Encoding for 'Embarked'

```python
# Using Pandas
pd.get_dummies(
    titanic_df, columns=['Embarked'], dtype=int
)
```

```python
# Using Scikit-learn
onehot_encoder = OneHotEncoder(sparse_output=False)
encoded_data = onehot_encoder.fit_transform(titanic_df[['Gender', 'Embarked']])
encoded_data
```
