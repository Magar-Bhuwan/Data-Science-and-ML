# apply, map, aggregate function

# apply
- DataFrame.apply(func, axis=0) calls func on each column by default. With axis=1, it calls func on each row. Series.apply(func) calls func on each element (or the whole Series, depending on signatures in recent versions).

```python
# apply

# def get_initial(name):
#     return name.split(",")[-1].split(".")[0].strip()

get_initial = lambda name: name.split(",")[-1].split(".")[0].strip()

titanic_df.Name.apply(get_initial)

```

```python
get_initial = lambda name: name.split(",")[-1].split(".")[0].strip()

titanic_df.Name.map(get_initial)
```

```python
get_max = lambda values: values.max()

titanic_df.Age.aggregate(["mean", "sum", get_max])
```

# Handling Incorrect Data

```python
import numpy as np


data = pd.DataFrame({
    "id":[1,3,5,7,9],
    "name":["John","Jane","Jennifer","Johan","Jeff"],
    "age":[25,-32,19,42,"55"],
    "gender" : ["M","M","Z","F","F"],
    "income":[400, 700, 300, np.nan, 600]
})
```

```python
data
```

```python
data.dtypes
```

```python
data = data.astype({"age":"int16"})
data.age = abs(data.age)
```
