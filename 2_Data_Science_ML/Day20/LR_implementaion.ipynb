```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

# EDA

```python
data.info()
```

```python
data.Outcome.value_counts().plot(kind = 'barh')
```

```python
data.plot(kind="scatter", x = "BMI", y="Outcome", c = "Pregnancies", cmap = "Dark2")
```

```python
x = data.drop("Outcome", axis = 1).to_numpy()
y = data[["Outcome"]].to_numpy()
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    scaled_X, y, test_size=0.2, random_state=999
)
```

```python
log_res = LogisticRegression()
log_res.fit(X_train, y_train)
```

```python
y_pred = log_res.predict(X_test)
```

```python
y_pred
```

```python
y_test.flatten()
```

```python
score = np.sum(y_pred == y_test) / y_test.shape[0]
score
```

```python
log_res.score(X_test, y_test)
```
