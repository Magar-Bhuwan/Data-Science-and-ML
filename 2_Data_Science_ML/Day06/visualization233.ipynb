```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Use for
Use Seaborn for better visuals than Matplotlib.

Always clean data before plotting.
- hist → distribution
- bar/countplot → comparison
- heatmap → relationships
- boxplot → outliers

```python
# Optional but better visualization
sns.set_style('whitegrid')
```

```python
# Load Dataset
titanic_df = pd.read_csv("titanic.csv")
```

```python
# Understand structure before visualization
titanic_df.head()
titanic_df.info()
titanic_df.describe()
```

```python
# Check missing value
titanic_df.isnull().sum()
```

```python
# Data Cleaning is important for accuracy
# Handling the missing values using fillna
titanic_df.loc[:, 'Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df.loc[:, 'Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
```

```python
# Count Plot (Survival Count)
sns.countplot(x='Survived', data=titanic_df)
plt.title("Survival Count")
plt.show()
```

```python
# Bar Plot (Survival by Gender)
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title("Survival by Gender")
plt.show()
```

```python
# Histogram (Age Distribution)
titanic_df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
# Box Plot (Age vs Survival)
sns.boxplot(x='Survived', y='Age', vert = True, data=titanic_df)
plt.title("Age vs Survival")
plt.show()
```

```python
# Heatmap (Correlation Matrix)
corr = titanic_df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Grouping + Visualization using groupby
grouped = titanic_df.groupby('Pclass')['Survived'].mean()
grouped.plot(kind='bar', title="Survival Rate by Class")
plt.show()
```

```python
# Pie Chart (Embarked Distribution)
titanic_df['Embarked'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    title="Passenger Distribution by Embarked"
)
plt.ylabel('')
plt.show()
```

```python
# Crosstab Visualization (Very Important)
pd.crosstab(
    titanic_df['Sex'],
    titanic_df['Survived'],
    normalize='index'
).plot(kind='bar', stacked=True)

plt.title("Survival Rate by Gender (Percentage)")
plt.show()
```
