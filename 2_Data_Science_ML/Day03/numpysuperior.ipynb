Why Numpy over python for loop?

```python
import numpy as np
import time
```

```python
a = np.random.rand(10_000_000)
b = np.random.rand(10_000_000)
```

```python
st = time.time()

result =[]
for num1, num2 in zip(list(a), list(b)):
    result.append(num1*num2)

end = time.time()

print("Time taken for list comprehension: ", end-st)
```

```python
# Comprehension

st = time.time()

result =[num1*num2 for num1, num2 in zip(list(a), list(b))]

end = time.time()

print("Time taken for list comprehension: ", end-st)
```

```python
# Numpy Operations

st = time.time()
result = np.multiply(a, b)
end = time.time()
print("Time taken for numpy multiplication: ", end-st)
```
