```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_data = pd.read_csv('tips.csv')
```

```python
tips_data
```

```python
tips_data.time.value_counts()
```

```python
sns.countplot(x = 'day',
              data = tips_data,
              hue = 'day',
              palette = 'pastel',
              legend=True
)
plt.show()
```

```python
day_counts = tips_data['day'].value_counts()

sns.barplot(
    x=day_counts.index,
    y=day_counts.values,
    hue=day_counts.index,
    palette="viridis",
    legend=False
)

plt.title("Count of Records per Day")
plt.show()
```

```python
count_info = tips_data.smoker.value_counts()
count_info
```

```python
count_info.plot(
    kind = 'bar',
    title = "Smoker and Non Smoker data",
    ylabel = "Total people", 
    color = ['green', 'red']

    )
```

```python
ax = pd.crosstab(
    tips_data['smoker'],
    tips_data['sex']
).plot(
    kind='bar',
    title="Smoker and Non Smoker data",
    ylabel="Total people",
    color=['orange', 'blue']
)

# Add numbers on top of bars
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
day_counts.plot(kind = 'pie',
                y = 'day',
                autopct = '%1.2f'
                )
plt.title("Tips Data by Days")
plt.show()
```

```python
smoker_values = tips_data[['smoker','sex']].value_counts()
smoker_values
```

```python
ax = sns.countplot(
    tips_data,
    x='smoker',
    hue='sex'
)
# Add numbers on top of bars
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
sns.scatterplot(
    data=tips_data,
    x = 'total_bill',
    y = 'tip',
    hue = 'day'
)
```

```python
sns.displot(tips_data['total_bill'])
```

```python
total_b = tips_data['total_bill']
mean = total_b.mean()
median = total_b.median()
mode = total_b.mode()[0]
```

```python
sns.displot(total_b, kde=True)

plt.axvline(mean, color='r', label='mean')
plt.axvline(median, color='g', label='median')
plt.axvline(mode, color='b', label='mode')

plt.legend()
plt.show()
```

Above chart shows the mean, median and mode of distribution of total bill of data.
It shows more concentrate data are placed in the between of 10 - 20 amount of bill and skewed to the right side.

```python
sns.boxplot(total_b,
            orient='h', 
            color='Brown')
plt.show()
```

Above boxplot shows the:
- Most bills are clustered in a certain range (middle box)
- A few very high bills exist (right-side outliers)
- Distribution is usually right-skewed (long tail to the right)

```python
sns.violinplot(total_b, color = 'orange', orient='h')
plt.show()
```

```python
total_b.kurt()
```

Kurtosis measures: how heavy the tails or outliers of a distribution are.

📊 Interpretation

Kurtosis Value	Meaning

- (≈ 0)	Normal distribution
- (> 0)	More outliers / heavy tails
- (< 0)	Flatter distribution / fewer outliers

```python
total_b.skew()
```

Skewness tells you: whether the distribution leans more to the left or right. (1.13 > 0) so, the data are positive right skewed.
